```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_excel('student_lifestyle.xlsx')
df.head()
```

```python
df.columns
```

```python
colunas = df[['Study_Hours_Per_Day', 'Grades']]
colunas.head()
```

```python
correlacao = colunas.corr()
print(correlacao)
```

```python
plt.scatter(df['Study_Hours_Per_Day'], df['Grades'], s=10, color='purple')
plt.xlabel('Horas de estudo diárias ')
plt.ylabel('Notas Finais')
plt.title('Correlação entre Horas estudadas e Notas Finais', fontsize=14, fontweight='bold')
plt.show()
```

```python
import seaborn as sns


# Selecionado apenas colunas numéricas
colunas_numericas = ['Study_Hours_Per_Day', 'Extracurricular_Hours_Per_Day',
                     'Sleep_Hours_Per_Day', 'Social_Hours_Per_Day',
                     'Physical_Activity_Hours_Per_Day', 'Grades']

correlation_matrix = df[colunas_numericas].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Purples', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlação entre Variáveis do Estilo de Vida', fontsize=14, fontweight='bold')
plt.show()
```
